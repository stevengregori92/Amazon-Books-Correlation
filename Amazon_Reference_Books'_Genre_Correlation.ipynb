#Import Libraries

```python
import pandas as pd
import seaborn as sns
```

#Import Dataset

```python
data = pd.read_csv('Reference.csv')
data.head()
```

```python
data.shape
```

#Explore Dataset

```python
len(data), len(data.dropna())
```

```python
data.info()
```

#Data Correlation

```python
sns.heatmap(data.corr(), annot = True)
```

#Data Visualization

```python
sns.regplot(x = 'five_star_rating', y = 'overall_rating', data = data)
```

```python
sns.regplot(x = 'three_star_rating', y = 'four_star_rating', data = data)
```

```python
sns.regplot(x = 'two_star_rating', y = 'three_star_rating', data = data)
```

```python
sns.regplot(x = 'one_star_rating', y = 'two_star_rating', data = data)
```

#Report Findings

Insights:
- Overall rating has 94% correlated with 5 star rating.
- Four star rating has 64% correlated with three star rating.
- three star rating has 77% correlated with two star rating.
- two star rating has 53% correlated with one star rating.

Conclution: Reference genre books' rating in Amazon mostly influenced with 5 star rating.
