#Import Libraries

```python
import pandas as pd
import seaborn as sns
```

#Import Dataset

```python
data = pd.read_csv('Religion (1).csv')
data.head()
```

```python
data.shape
```

#Explore Dataset

```python
len(data), len(data.dropna())
```

```python
data.info()
```

#Data Correlation

```python
sns.heatmap(data.corr(), annot = True)
```

#Data Visualization

```python
sns.regplot(data = data, x = 'overall_rating', y = 'five_star_rating')
```

```python
sns.regplot(data = data, x = 'four_star_rating', y = 'three_star_rating')
```

```python
sns.regplot(data = data, x = 'three_star_rating', y = 'two_star_rating')
```

```python
sns.regplot(data = data, x = 'two_star_rating', y = 'one_star_rating')
```

#Report Findings

Insights:
- Overall rating has 93% correlated with 5 star rating.
- Four star rating has 75% correlated with three star rating.
- three star rating has 78% correlated with two star rating.
- two star rating has 67% correlated with one star rating.

Conclution:
- Religion genre books' overall rating in Amazon really influenced by 5 star rating.
- Each of every star ratings (example 4 star with 3 star, 3 star with 2 star, etc) has a really high correlation
